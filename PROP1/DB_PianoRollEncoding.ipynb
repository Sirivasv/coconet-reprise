```python
# Traverse midi files
nes_song_filenames = []
for root, directories, files in os.walk(NES_DATASET_PATH):
    for file in files:
        nes_song_filenames.append(file)
print(nes_song_filenames[:3])
print(len(nes_song_filenames))
```

```python
%%time

# Read Files 
MAX_LIM_NES_SONGS = 2500
len_nes_song_filenames = len(nes_song_filenames)
nes_songs_with_error = []
nes_songs_NO_error = []
nes_songs_NO_error_data = []


# We traverse the reduced table
cnt = 1
for song_data_test in sorted_list_LOADED["data"]:
    song_filename_test = song_data_test[0]
    try:
        
        # midi_melody_file = pretty_midi.PrettyMIDI(os.path.join(NES_DATASET_PATH, song_filename_query))
        midi_fullvoices_file = pretty_midi.PrettyMIDI(os.path.join(NES_DATASET_PATH, song_filename_test))

        # pianoroll_melody_result = encoder_decoder.encode_midi_melody_to_pianoroll(midi_melody_file)
        pianoroll_fullvoices_result = encoder_decoder.encode_midi_to_pianoroll(midi_fullvoices_file, (1, 32, 46, 4))
        
        nes_songs_NO_error.append(pianoroll_fullvoices_result[0])
        nes_songs_NO_error_data.append(song_data_test)
        
    except:
        print("[ERROR!]")
        nes_songs_with_error.append(song_filename_test)
    finally:
        print("{0}/{1} - {2}".format(cnt, MAX_LIM_NES_SONGS, song_filename_test))
        cnt += 1
        if (cnt == MAX_LIM_NES_SONGS):
            break
```
